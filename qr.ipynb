```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import qr
# unshift QR algorithm:
#using PyPlot

#make a random matrix
array = np.random.randint(10, size=(100, 100))

# #symmetriz it
symm=np.matmul(array,array.T)
plt.imshow(symm)
# #A = rand(100,100); 
# A=array
# #perform QR iteration
for k in range(2):
    (Q,R)=qr(symm)
    symm=R*Q

# #spy(abs(A).>1e-4)
```

```python
#plot it
plt.imshow(np.log10(abs(symm))>1.e-4)
```
